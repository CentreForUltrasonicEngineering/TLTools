```python
%matplotlib inline
from TLTools import Imaging
import scipy.io as sio

#FMC_file = 'TestData.mat'
FMC_file = 'M2.mat'
FMC = sio.loadmat(FMC_file)
FMC_dataset =  FMC['FMC'].T
#ProbeElementLocations = FMC['Array']

TFM = Imaging.PyTFM()
```

```python
TFM.uploadFMC(FMC_dataset)
#TFM.uploadProbe(ProbeElementLocations)
TFM.buildProbe(0.7e-3)
TFM.setParameters(Velocity1=5790,Fs=72792400,Ts=0)
TFM.setRefraction(RefractionType='flat',Velocity2=1496)
TFM.raiseArrayToHeight(2.895e-3)
TFM.setImage(y0=-30e-3,ny=1024,y1=30e-3,z0=0,nz=1024,z1=-30e-3)
TFM.printTFM(range=50)
```
